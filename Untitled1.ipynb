```python
# Importando as ferramentas antes de tudo
import pandas as pd
import numpy as np

# ---------------------------------------- PARTE 1: A RECEITA DO BOLO (FUNÇÃO) ----------------------------------------
# def: "Estou criando uma nova receita"
# criar_dados_loja: "É o nome que demos à receita"
# ( ... ): "Dentro dos parênteses estão os ingredientes que nossa receita precisa:
# n_vendas, preco_medio, avaliacao_media, frete_medio, nome_loja"

def criar_dados_loja(n_vendas, preco_medio, avaliacao_media, frete_medio, nome_loja):

    # ---------------------------------------- PARTE 2: O QUE ACONTECE DENTRO DA FÁBRICA ----------------------------------------

    # Criamos uma lista com todos os tipos de categorias de produtos que uma loja pode vender
    categorias = ['eletronicos', 'moveis', 'livros', 'esporte e lazer', 'cozinha', 'vestuario']

    # Criamos um dicionário, como uma ficha: cada item tem um título (a chave)
    # e um preenchimento (o valor)
    data = {
        # Campo 1: Nome do Produto
        'Produto': [f'Produto {i}' for i in range(n_vendas)],

        # Campo 2: Categoria do Produto
        # Usamos o np.random.choice para escolher aleatoriamente categorias
        # com chances diferentes usando p=[...].
        'Categoria do Produto': np.random.choice(
            categorias, n_vendas, p=[0.3, 0.2, 0.1, 0.1, 0.2, 0.1]
        ),

        # Campo 3: Preço de cada venda
        # Geramos preços em torno do preco_medio para criar variações naturais
        'Preço': np.random.normal(loc=preco_medio, scale=preco_medio * 0.5, size=n_vendas).clip(min=10),

        # Campo 4: Frete
        'Frete': np.random.normal(loc=frete_medio, scale=frete_medio * 0.3, size=n_vendas).clip(min=0),

        # Campo 5: Avaliação da compra
        # Gera notas em torno da avaliacao_media, limita entre 1 e 5 e transforma em inteiro
        'Avaliação da compra': np.random.normal(
            loc=avaliacao_media, scale=0.8, size=n_vendas
        ).clip(1, 5).astype(int),
    }

    # return indica: “Este é o resultado final”
    # Usamos pd.DataFrame(data) para transformar a ficha em uma tabela organizada
    return pd.DataFrame(data)

# ---------------------------------------- PARTE 3: USANDO A FÁBRICA PARA CRIAR AS LOJAS ----------------------------------------
print("Criando os dados para TODAS as 4 lojas...")

# --> CORREÇÃO: ADICIONAMOS A CRIAÇÃO DAS LOJAS 1, 2 E 3 QUE ESTAVAM FALTANDO.
loja1 = criar_dados_loja(n_vendas=2500, preco_medio=1300, avaliacao_media=4.2, frete_medio=25, nome_loja='Loja 1')
loja2 = criar_dados_loja(n_vendas=3000, preco_medio=1500, avaliacao_media=4.0, frete_medio=45, nome_loja='Loja 2')
loja3 = criar_dados_loja(n_vendas=1800, preco_medio=900, avaliacao_media=4.8, frete_medio=20, nome_loja='Loja 3')

# Essa linha que você já tinha agora funciona em conjunto com as outras
loja4 = criar_dados_loja(n_vendas=1500, preco_medio=800, avaliacao_media=3.1, frete_medio=35, nome_loja='Loja 4')

print("Dados das 4 lojas criados com sucesso!")

# ---------------------------------------- PASSO 4: CALCULANDO AS MÉTRICAS ----------------------------------------
# --> Agora esta parte vai funcionar, porque as variáveis loja1, loja2 e loja3 existem!

# Juntamos os dados e nomes das lojas em listas para facilitar a análise
dataframes = [loja1, loja2, loja3, loja4]
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

print("\n--- Análise de Desempenho das Lojas ---")

# Loop que percorre cada loja, calcula as métricas e imprime o resumo
for i, df in enumerate(dataframes):
    loja_nome = nomes_lojas[i]

    # 1. Faturamento total (soma dos preços)
    faturamento = df['Preço'].sum()

    # 2. Avaliação média
    avaliacao_media = df['Avaliação da compra'].mean()

    # 3. Frete médio
    frete_medio = df['Frete'].mean()

    # Exibe os resultados formatados
    print(f"\n{loja_nome}:")
    print(f"  Faturamento Total: R$ {faturamento:,.2f}")
    print(f"  Avaliação Média: {avaliacao_media:.2f} de 5")
    print(f"  Frete Médio: R$ {frete_medio:.2f}")
```

```python
# Gráfico 1: Comparando o faturamento

# Primeiro, criamos uma lista com os faturamentos de cada loja
faturamentos = [
    loja1['Preço'].sum(),
    loja2['Preço'].sum(),
    loja3['Preço'].sum(),
    loja4['Preço'].sum()
]
# E uma lista com os nomes das lojas
lojas_nomes = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# plt.figure define o tamanho da imagem do nosso gráfico
plt.figure(figsize=(10, 6))

# plt.bar é o comando que cria o gráfico de barras
plt.bar(lojas_nomes, faturamentos, color=['#4285F4', '#DB4437', '#F4B400', '#0F9D58'])

# Adicionamos um título e nomes para os eixos para o gráfico ficar claro
plt.title('Faturamento Total por Loja', fontsize=16)
plt.ylabel('Faturamento (em milhões de R$)')
plt.xlabel('Lojas')

# plt.show() exibe o gráfico final
plt.show()
```

```python
# Gráfico 2: Comparando a satisfação dos clientes

# Criamos uma lista com as avaliações médias de cada loja
avaliacoes = [
    loja1['Avaliação da compra'].mean(),
    loja2['Avaliação da compra'].mean(),
    loja3['Avaliação da compra'].mean(),
    loja4['Avaliação da compra'].mean()
]
# E a lista com os nomes das lojas novamente
lojas_nomes = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(10, 6))

# plt.barh cria um gráfico de barras na HORIZONTAL
plt.barh(lojas_nomes, avaliacoes, color=['#4285F4', '#DB4437', '#F4B400', '#0F9D58'])

# Adicionamos os títulos
plt.title('Avaliação Média dos Clientes por Loja', fontsize=16)
plt.xlabel('Avaliação Média (de 1 a 5)')

# Definimos o limite do eixo X para ir de 1 a 5, que é a escala de notas
plt.xlim(1, 5)

# Mostramos o gráfico
plt.show()
```
